```python
import numpy as np
import PIL
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
from utils import default_paths
info_fn = os.path.join(default_paths.stim_root, 'Indep_set_info.csv')
print('\nloading image info from %s\n'%info_fn)
info_df = pd.read_csv(info_fn, index_col=0)
```

```python
image_dir = os.path.join('/user_data/mmhender/stimuli/texture_synth_images')
important_layers = ['relu1_1', 'pool1','pool2','pool3','pool4']
     
```

```python
info_df['filename_raw'][0]
```

```python
plt.figure(figsize=(14,14))

ims_plot = np.arange(6)

pi=0;

for ii in ims_plot:
    
    target_image_filename = info_df['filename_raw'].iloc[ii]
    target_image = utilities.preprocess_image_simple(
            utilities.load_image(target_image_filename)
        )
    pi+=1
    plt.subplot(len(ims_plot), len(important_layers)+1, pi)
    plt.imshow(target_image)
    plt.axis('off')
    if ii==0:
        plt.title('target')
        
    cocoid = '%012.f'%info_df['cocoId'][ii]
    
    folder = os.path.join(image_dir, 'coco_%s'%cocoid)
    
    for ll in range(len(important_layers)):
          
        filename_full = os.path.join(folder, 'upto_%s.png'%important_layers[ll])
        
        im = PIL.Image.open(filename_full)
        
        pi+=1
        plt.subplot(len(ims_plot), len(important_layers)+1, pi)
        plt.imshow(im)
        plt.axis('off')
        if ii==0:
            plt.title('match up to %s'%important_layers[ll])
       
```
