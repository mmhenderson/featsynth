```python
import numpy as np
import PIL
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
nsd_stim_path = '/user_data/mmhender/nsd/stimuli/'

info_fn = os.path.join(nsd_stim_path, 'Indep_set_info.csv')
print('\nloading image info from %s\n'%info_fn)
info_df = pd.read_csv(info_fn, index_col=0)
```

```python
n_ims_do=100
# n_ims_do = 16

filenames = target_image_filename = info_df['filename_raw'].iloc[np.arange(n_ims_do)]
cocoids = [fn.split('/')[-1].split('.')[0] for fn in filenames]
       
```

```python
image_dir = os.path.join('/user_data/mmhender/stimuli/texture_synth_images_grid2')
important_layers = ['relu1_1', 'pool1','pool2','pool3','pool4']
     
```

```python
# ims2plot = [4,5,6,7,8,9]
# ims2plot = [10,11,12,13,14,15]
ims2plot = np.arange(20,27)
n_ims_plot = len(ims2plot)

for n_grid in [1,2,4]:


    plt.figure(figsize=(14,12))

    pi = 0;

    for ii, im in enumerate(ims2plot):

        cid = cocoids[im]

        folder = os.path.join(image_dir, 'coco_%s'%cid)

        filename_full = os.path.join(folder, 'orig.png')
        im = PIL.Image.open(filename_full)

        pi+=1
        plt.subplot(n_ims_plot, len(important_layers)+1, pi)
        plt.imshow(im)
        plt.axis('off')
        if ii==0:
            plt.title('target')

        for ll in range(len(important_layers)):

            filename_full = os.path.join(folder, 'grid%dx%d_upto_%s.png'%\
                                         (n_grid, n_grid,important_layers[ll]))
            # print(filename_full)
            im = PIL.Image.open(filename_full)

            pi+=1
            plt.subplot(n_ims_plot, len(important_layers)+1, pi)
            plt.imshow(im)
            plt.axis('off')
            if ii==0:
                plt.title('%s'%important_layers[ll])

    plt.suptitle('synths with %d x %d grid'%(n_grid, n_grid));
```
